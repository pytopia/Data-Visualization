# Figure and Axes Objects

Welcome to our lecture on Figure and Axes objects in Matplotlib! Understanding these fundamental components is crucial for creating and manipulating plots effectively. In this session, we'll explore the structure of Matplotlib plots and learn how to work with Figures and Axes.


In Matplotlib, plots are organized in a hierarchical structure, with two main components at the core: **Figure** and **Axes**.


A **Figure** is the top-level container in Matplotlib. It represents the entire window or page where your plot(s) will be drawn. Think of it as a canvas that can contain one or more plots.


Key points about Figures:
- A Figure can hold multiple Axes (individual plots).
- It controls the overall size and dpi (dots per inch) of the plotting area.
- Figures can have additional elements like titles, legends, and colorbars that apply to all contained plots.


An **Axes** is the actual plotting area where data is visualized. It's where you'll find your plotted data, axis labels, tick marks, and titles.


Key points about Axes:
- Each Axes represents a single plot within a Figure.
- It contains methods for plotting various types of data (e.g., `plot()`, `scatter()`, `bar()`).
- Axes control aspects like axis limits, scales, and labels.
- A Figure can contain multiple Axes, allowing for subplots.


To visualize this relationship:

```
Figure
└── Axes
    ├── Data (Line, Points, Bars, etc.)
    ├── X-axis (with ticks and label)
    ├── Y-axis (with ticks and label)
    └── Title
```


Understanding this hierarchy is crucial because it determines how you create, customize, and manage your plots. For instance:
- Some methods are applied to the Figure (like `fig.suptitle()` for an overall title).
- Other methods are applied to specific Axes (like `ax.set_xlabel()` for an x-axis label).


In the following sections, we'll dive deeper into creating and manipulating Figures and Axes, giving you the tools to craft sophisticated and informative visualizations.

## <a id='toc1_'></a>[Creating a Figure](#toc0_)

In Matplotlib, creating a Figure is typically the first step in generating a visualization. Let's explore how to create and customize a Figure object.


### <a id='toc1_1_'></a>[Basic Figure Creation](#toc0_)


The simplest way to create a Figure is using the `plt.figure()` function:


````python
%matplotlib inline
````

````python
import matplotlib.pyplot as plt

plt.figure()
````

This creates an empty Figure object with default size and properties.


### <a id='toc1_2_'></a>[Customizing Figure Creation](#toc0_)


You can customize your Figure by passing parameters to `plt.figure()`:


````python
plt.figure(figsize=(10, 6), dpi=100, facecolor='lightblue', edgecolor='red')
````

Let's break down these parameters:

- `figsize`: A tuple (width, height) in inches. Default is typically (6.4, 4.8).
- `dpi`: Dots per inch. Higher values result in larger images. Default is usually 100.
- `facecolor`: Background color of the Figure. Can be any valid color format.
- `edgecolor`: Color of the Figure edge. Can be any valid color format.


### <a id='toc1_3_'></a>[Other Useful Parameters](#toc0_)


- `num`: An identifier for the Figure. Can be an integer or a string.
- `frameon`: If False, suppresses drawing the Figure frame.
- `tight_layout`: Automatically adjusts subplot params for a tight layout.


Example with additional parameters:


````python
plt.figure(num="My Custom Figure", frameon=False, tight_layout=True)
````

### <a id='toc1_4_'></a>[Displaying the Figure](#toc0_)


In a Jupyter notebook with `%matplotlib inline`, the Figure will display automatically. In a script, you might need to call `plt.show()` to display the Figure.


In Jupyter notebooks using Matplotlib, the display of figures is handled automatically through a mechanism called the "inline backend." Here's a brief explanation of how it works:

1. Jupyter integration: Matplotlib has built-in support for Jupyter notebooks.

2. IPython magic: When you use the `%matplotlib inline` magic command, it sets up the inline backend.

3. Automatic display: The inline backend automatically displays figures at the end of each cell execution.

4. Figure object tracking: Matplotlib keeps track of figure objects created in each cell.

5. Cell execution completion: When a cell finishes executing, Matplotlib checks for any newly created or modified figures.

6. Display trigger: If new or modified figures are found, they are automatically rendered and displayed below the cell.

7. Clearing: After display, the figures are typically cleared to prevent duplicate displays in subsequent cells.

This automatic display mechanism makes it convenient to work with plots in Jupyter notebooks without explicitly calling `plt.show()` or other display functions.

Remember, creating a Figure is just the first step. To actually plot data, you'll need to add Axes to your Figure, which we'll cover in the next section.


For a comprehensive list of Figure parameters and methods, refer to the [Matplotlib Figure documentation](https://matplotlib.org/stable/api/figure_api.html). This resource provides detailed information on all available options for fine-tuning your Figures.

## <a id='toc2_'></a>[Understanding Axes](#toc0_)

Axes are the heart of Matplotlib plots. They contain the data, axes lines, labels, and titles that make up a plot. Let's dive into how to create, manipulate, and understand Axes objects.


### <a id='toc2_1_'></a>[Creating Axes](#toc0_)


The most common way to create Axes is by using the `add_subplot()` method of a Figure object:


````python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
````

Here, `111` is a shorthand that means "1x1 grid, first subplot". It's equivalent to `(1, 1, 1)`.


You can also use plt.subplots() to create a Figure and Axes at the same time:

````python
fig, ax = plt.subplots(figsize=(6, 4))
````

### <a id='toc2_2_'></a>[Anatomy of Axes](#toc0_)


An Axes object contains several key components:

- **Data Area**: Where your plot is displayed
- **X and Y axes**: The lines that frame your data
- **Ticks**: The markers along the axes
- **Labels**: Text describing the axes and data
- **Title**: A text description of the plot


### <a id='toc2_3_'></a>[Adding Data to Axes](#toc0_)


Once you have an Axes object, you can add data to it:


````python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y)
````

````python
fig
````

### <a id='toc2_4_'></a>[Customizing Axes](#toc0_)


Axes objects have numerous methods for customization:


````python
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Sine Wave')
ax.grid(True)
fig
````

### <a id='toc2_5_'></a>[Axes vs. `plt.gca()`](#toc0_)


When using the pyplot interface, `plt.gca()` (get current axes) returns the current Axes object and `plt.gcf()` (get current figure) returns the current Figure object. These functions are useful for accessing and modifying the current plot. However, many pyplot functions implicitly create a Figure and Axes if they don't exist:

````python
plt.plot(x, y)  # Implicitly creates Figure and Axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sine Wave')
plt.grid(True)
````

While convenient, this implicit creation can sometimes lead to unexpected behavior in more complex scripts.


For a comprehensive list of Axes methods and properties, refer to the [Matplotlib Axes documentation](https://matplotlib.org/stable/api/axes_api.html). This resource provides detailed information on all available options for customizing your plots through Axes objects.


Understanding Axes is crucial for creating effective visualizations with Matplotlib. In the next section, we'll explore how to work with multiple Axes in a single Figure using subplots.

## <a id='toc3_'></a>[Subplots: Multiple Axes in a Figure](#toc0_)

Subplots allow you to create multiple plots within a single Figure, enabling side-by-side comparisons or the presentation of related data. Let's explore how to create and manipulate subplots in Matplotlib.


### <a id='toc3_1_'></a>[Creating Subplots with plt.subplots()](#toc0_)


The most convenient way to create subplots is using `plt.subplots()`:


````python
fig, axs = plt.subplots(2, 2, figsize=(8, 4))
````

This creates a 2x2 grid of subplots. `fig` is the Figure object, and `axs` is a 2D array of Axes objects.


### <a id='toc3_2_'></a>[Adding Data to Subplots](#toc0_)


You can access individual Axes using indexing:


````python
x = np.linspace(0, 10, 100)

axs[0, 0].plot(x, np.sin(x))
axs[0, 1].plot(x, np.cos(x))
axs[1, 0].plot(x, np.exp(x))
axs[1, 1].plot(x, np.log(x))

fig
````

### <a id='toc3_3_'></a>[Customizing Individual Subplots](#toc0_)


Each Axes can be customized independently:


````python
axs[0, 0].set_title('Sine')
axs[0, 1].set_title('Cosine')
axs[1, 0].set_title('Exponential')
axs[1, 1].set_title('Logarithm')

for ax in axs.flat:
    ax.set(xlabel='x-axis', ylabel='y-axis')
    ax.grid(True)

# set sharex to True to share the x-axis
plt.subplots(2, 2, figsize=(8, 4), sharex=True)
````

### <a id='toc3_4_'></a>[Adjusting Subplot Layouts](#toc0_)


Matplotlib provides tools to adjust the layout of subplots:


````python
fig, axs = plt.subplots(2, 2, figsize=(8, 4))
fig.tight_layout(pad=0)
````

````python
fig.tight_layout(pad=10)
fig
````

`tight_layout()` automatically adjusts subplot params for a tight fit.


### <a id='toc3_5_'></a>[Creating Subplots with Unequal Sizes](#toc0_)


You can create subplots of different sizes using `gridspec`:


````python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(6, 4))
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0, :])  # Top row, span all columns
ax2 = fig.add_subplot(gs[1, 0])  # Bottom left
ax3 = fig.add_subplot(gs[1, 1])  # Bottom right

ax1.plot(x, np.sin(x))
ax2.plot(x, np.exp(x))
ax3.plot(x, np.log(x))
````

### <a id='toc3_6_'></a>[Sharing Axes](#toc0_)


You can share x or y axes among subplots:


````python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(6, 4))

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
````

### <a id='toc3_7_'></a>[Subplots with plt.subplot()](#toc0_)


For more control, you can use `plt.subplot()`:


````python
fig = plt.figure(figsize=(6, 4))

plt.subplot(221)
plt.plot(x, np.sin(x))

plt.subplot(222)
plt.plot(x, np.cos(x))

plt.subplot(212)
plt.plot(x, np.tan(x))
````

Here, `221` means "2x2 grid, first subplot", `222` means "2x2 grid, second subplot", and `212` means "2x1 grid, second subplot".


For more advanced subplot configurations and options, refer to the [Matplotlib Subplots documentation](https://matplotlib.org/stable/api/subplots_api.html). This resource provides detailed information on creating complex subplot layouts and customizing their behavior.


Mastering subplots allows you to create rich, informative visualizations that convey multiple aspects of your data in a single figure. In the next section, we'll explore how to further customize the appearance of your figures and axes.

## <a id='toc4_'></a>[Customizing Figure and Axes Properties](#toc0_)

Customizing your Figure and Axes is crucial for creating clear, informative, and visually appealing plots. Let's explore various ways to modify the properties of these objects.


### <a id='toc4_1_'></a>[Figure Customization](#toc0_)


You can adjust the **size and resolution of your Figure**:


````python
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.plot([1, 2, 3, 4])
````

Change the **Figure's background color**:


````python
fig, ax = plt.subplots()
fig.patch.set_facecolor('#E6E6E6')
ax.plot([1, 2, 3, 4])
fig
````

Add **titles** to your Figure:


````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))
ax1.plot([1, 2, 3, 4])
ax2.plot([4, 3, 2, 1])

fig.suptitle('Main Title', fontsize=16)
ax1.set_title('Subplot 1')
ax2.set_title('Subplot 2')
````

### <a id='toc4_2_'></a>[Axes Customization](#toc0_)


Set **labels and titles** for your Axes:


````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Plot Title')
````

Control the **range of your axes**:


````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])

ax.set_xlim(0, 5)
ax.set_ylim(0, 6)
````

Add or customize **grid lines**:


````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])

ax.grid(True, linestyle='--', alpha=0.7)
````

Customize **tick marks and labels**:


````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])

ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(['Zero', 'One', 'Two', 'Three', 'Four'], rotation=45)
ax.set_yticks([1, 2, 3, 4])
ax.set_yticklabels(['Low', 'Medium', 'High', 'Very High'])
````

### <a id='toc4_3_'></a>[Styling with rcParams](#toc0_)


For consistent styling across all plots, you can modify Matplotlib's **rcParams**. This stands for "Run Control" or "Runtime Configuration." It's a common convention in Unix-like systems for configuration files.

````python
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['figure.titlesize'] = 20

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], label='Data')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Plot with Custom Style')
ax.legend()
````

### <a id='toc4_4_'></a>[Using Style Sheets](#toc0_)


Matplotlib provides pre-defined style sheets for quick styling:


````python
plt.style.use('seaborn-v0_8-darkgrid')

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])
ax.set_title('Plot with Seaborn Style')
````

You can see a list of available styles with `plt.style.available`:

````python
plt.style.available
````

To reset all rcParams to their default values:

````python
plt.rcParams.update(plt.rcParamsDefault)
````

For a comprehensive list of customizable properties and methods, refer to the [Matplotlib Figure](https://matplotlib.org/stable/api/figure_api.html) and [Axes](https://matplotlib.org/stable/api/axes_api.html) documentation. These resources provide detailed information on all available options for fine-tuning your plots.

### <a id='toc5_1_'></a>[Example 1: Multi-line Plot with Legend](#toc0_)


Let's create a plot comparing multiple datasets:


````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/10)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y1, label='Sin(x)', color='blue')
ax.plot(x, y2, label='Cos(x)', color='red')
ax.plot(x, y3, label='Exp(-x/10)', color='green')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Comparison of Functions')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

fig.tight_layout()
````

This example shows how to plot multiple lines, add a legend, customize colors, and add grid lines.


### <a id='toc5_2_'></a>[Example 2: Scatter Plot with Color Mapping](#toc0_)


Let's create a scatter plot with points colored based on a third variable:


````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 200 * np.random.rand(100)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))

scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot with Color Mapping')

# Add a color bar
cbar = fig.colorbar(scatter)
cbar.set_label('Color Value')

fig.tight_layout()
````

This example demonstrates creating a scatter plot with points sized and colored based on data values, and how to add a colorbar.


### <a id='toc5_3_'></a>[Example 3: Subplots with Shared Axis](#toc0_)


Let's create multiple subplots with a shared x-axis:


````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create the plot
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 4), sharex=True)

ax1.plot(x, y1, color='blue')
ax1.set_title('Sin(x)')
ax1.set_ylabel('Amplitude')

ax2.plot(x, y2, color='red')
ax2.set_title('Cos(x)')
ax2.set_ylabel('Amplitude')

ax3.plot(x, y3, color='green')
ax3.set_title('Tan(x)')
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Amplitude')

# Adjust y-limits for tan(x) plot
ax3.set_ylim(-5, 5)

fig.suptitle('Trigonometric Functions', fontsize=16)
fig.tight_layout()
````

This example shows how to create subplots with a shared x-axis, customize individual subplots, and add an overall title to the figure.


### <a id='toc5_4_'></a>[Example 4: Bar Plot with Error Bars](#toc0_)


Let's create a bar plot with error bars:


````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 50, size=5)
errors = np.random.randint(1, 5, size=5)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))

ax.bar(categories, values, yerr=errors, capsize=5, 
       color='skyblue', edgecolor='navy')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar Plot with Error Bars')

# Add value labels on top of each bar
for i, v in enumerate(values):
    ax.text(i, v + errors[i] + 0.5, str(v), ha='center')

ax.set_ylim(0, max(values + errors) * 1.2)  # Adjust y-axis limit

fig.tight_layout()
````

This example demonstrates creating a bar plot with error bars, customizing colors, and adding value labels to each bar.
