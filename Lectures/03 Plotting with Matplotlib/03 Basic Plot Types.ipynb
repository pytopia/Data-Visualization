# Basic Plot Types in Matplotlib

Welcome to our lecture on basic plot types in Matplotlib! In this session, we'll explore four fundamental visualization techniques that form the backbone of data representation: line plots, scatter plots, bar charts, and histograms.


Data visualization is a crucial skill in data analysis and scientific communication. It allows us to represent complex data in a visually appealing and easily understandable format. Matplotlib offers a wide range of plot types, but mastering these four basic types will give you a solid foundation for most data visualization tasks.

1. **Line Plots**: Ideal for showing trends over time or continuous data. They're perfect for visualizing relationships between two variables, especially when one is dependent on the other.

2. **Scatter Plots**: Excellent for displaying the relationship between two variables. They're particularly useful for identifying correlations, clusters, or outliers in your data.

3. **Bar Charts**: Great for comparing quantities across different categories. They're commonly used in business, economics, and social sciences to show comparative data.

4. **Histograms**: Used to show the distribution of a single variable. They're invaluable for understanding the shape, central tendency, and spread of your data.


Each of these plot types has its strengths and is suited to different kinds of data and analysis goals. By the end of this lecture, you'll be able to:

- Create each of these plot types using Matplotlib
- Customize their appearance to effectively communicate your data
- Choose the appropriate plot type for your specific data and analysis needs


Remember, the key to effective data visualization is not just creating the plot, but choosing the right type of plot that best represents your data and communicates your message clearly.


In the following sections, we'll dive deep into each of these plot types, exploring how to create them, customize them, and when to use them. Let's start our journey into the world of Matplotlib's basic plot types!

## <a id='toc1_'></a>[Line Plots](#toc0_)

Line plots are one of the most common and versatile plot types in data visualization. They're excellent for showing trends over time or relationships between continuous variables.


Let's start by creating a simple line plot:


```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(6, 3))

# Plot the data
ax.plot(x, y)

# Add title and labels
ax.set_title("Sine Wave")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
```

In this example, we:
1. Create our data using NumPy.
2. Use `plt.subplots()` to create a figure and axis object.
3. Use the `ax.plot()` method to create the line plot.
4. Set the title and axis labels using axis methods.


### <a id='toc1_1_'></a>[Customizing Line Plots](#toc0_)


We can customize various aspects of our line plot:


```python
fig, ax = plt.subplots(figsize=(7, 3))

# Plot multiple lines
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')

# Customize line styles
ax.plot(x, np.sin(x) + 2, linestyle='--', color='r', linewidth=2, label='sin(x) + 2')

# Add legend
ax.legend()

# Customize axes
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 3)

# Add grid
ax.grid(True)
```

This example demonstrates:
- Plotting multiple lines on the same axes
- Customizing line styles, colors, and widths
- Adding a legend
- Setting axis limits
- Adding a grid


### <a id='toc1_2_'></a>[When to Use Line Plots](#toc0_)


Line plots are ideal when you want to:
- Show trends over time
- Display continuous data
- Compare multiple series of data
- Visualize mathematical functions


In the next sections, we'll explore other plot types and see how they can be created and customized using the same `fig, ax = plt.subplots()` approach.

## <a id='toc2_'></a>[Scatter Plots](#toc0_)

Scatter plots are excellent for visualizing the relationship between two variables. They're particularly useful for identifying correlations, clusters, or outliers in your data.


Let's start with a simple scatter plot:


```python
# Create data
np.random.seed(42)
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(6, 3))

# Create the scatter plot
ax.scatter(x, y)

# Add title and labels
ax.set_title("Basic Scatter Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
```

In this example, we:
1. Generate random data using NumPy.
2. Create a figure and axis object using `plt.subplots()`.
3. Use `ax.scatter()` to create the scatter plot.
4. Set the title and axis labels.


### <a id='toc2_1_'></a>[Customizing Scatter Plots](#toc0_)


Scatter plots can be customized in various ways to enhance their informativeness:


```python
# Create data
np.random.seed(42)
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)
colors = np.random.rand(50)
sizes = 300 * np.random.rand(50)

fig, ax = plt.subplots(figsize=(6, 3))

# Create a scatter plot with customized points
scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')

# Add a color bar
cbar = fig.colorbar(scatter)
cbar.set_label('Color Value')

# Customize the plot
ax.set_title("Customized Scatter Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.grid(True)

# Add a trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax.plot(x, p(x), "r--", alpha=0.8, label="Trend Line")

ax.legend()
```

This example demonstrates:
- Varying point colors based on a third variable
- Adjusting point sizes
- Adding transparency with `alpha`
- Including a colorbar
- Adding a trend line
- Customizing the overall appearance with a grid and legend


### <a id='toc2_2_'></a>[When to Use Scatter Plots](#toc0_)


Scatter plots are ideal when you want to:
- Explore relationships between two continuous variables
- Identify correlations (positive, negative, or no correlation)
- Detect clusters or groupings in your data
- Spot outliers or unusual patterns
- Visualize multivariate data (using color and size for additional dimensions)


### <a id='toc2_3_'></a>[Scatter Plot Variations](#toc0_)


There are several variations of scatter plots that can be useful in specific scenarios:

1. **Bubble charts**: Scatter plots where point sizes represent a third variable.
2. **Pair plots**: A grid of scatter plots showing pairwise relationships in a dataset.
3. **3D scatter plots**: For visualizing relationships among three variables.

## <a id='toc3_'></a>[Bar Charts](#toc0_)

Bar charts are excellent for comparing quantities across different categories. They're widely used in business, economics, and social sciences to display comparative data.


### <a id='toc3_1_'></a>[Creating a Basic Bar Chart](#toc0_)


Let's start with a simple bar chart:


```python
# Create data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 35, 14, 27, 10]

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(7, 5))

# Create the bar chart
ax.bar(categories, values)

# Add title and labels
ax.set_title("Basic Bar Chart")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")

# Add value labels on top of each bar
for i, v in enumerate(values):
    ax.text(i, v, str(v), ha='center', va='bottom')
```

In this example, we:
1. Define categories and their corresponding values.
2. Create a figure and axis object using `plt.subplots()`.
3. Use `ax.bar()` to create the bar chart.
4. Set the title and axis labels.
5. Add value labels on top of each bar.


### <a id='toc3_2_'></a>[Customizing Bar Charts](#toc0_)


Bar charts can be customized in various ways to enhance their visual appeal and informativeness:


```python
# Create data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 35, 14, 27, 10]
values2 = [15, 29, 22, 18, 7]

fig, ax = plt.subplots(figsize=(12, 7))

# Create grouped bar chart
x = np.arange(len(categories))
width = 0.35

rects1 = ax.bar(x - width/2, values1, width, label='Group 1', color='skyblue')
rects2 = ax.bar(x + width/2, values2, width, label='Group 2', color='lightgreen')

# Customize the plot
ax.set_title("Customized Grouped Bar Chart")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Add a grid
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

# Add value labels
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
```

This example demonstrates:
- Creating a grouped bar chart to compare two sets of data
- Customizing bar colors
- Adding a legend
- Including a grid for easier value comparison
- Adding value labels to each bar
- Adjusting the overall layout and appearance


### <a id='toc3_3_'></a>[When to Use Bar Charts](#toc0_)


Bar charts are ideal when you want to:
- Compare quantities across different categories
- Show the distribution of data across discrete groups
- Display data that has distinct, separate categories
- Visualize survey results or categorical data

There are several variations of bar charts that can be useful in specific scenarios:

1. **Stacked bar charts**: For showing the composition of each category
2. **Horizontal bar charts**: Useful when you have long category names or many categories
3. **Error bar charts**: To display uncertainty or variability in the data

## <a id='toc4_'></a>[Histograms](#toc0_)

Histograms are powerful tools for visualizing the distribution of a single continuous variable. They provide insights into the central tendency, spread, and shape of your data.


### <a id='toc4_1_'></a>[Creating a Basic Histogram](#toc0_)


Let's start with a simple histogram:


```python
# Create data
np.random.seed(42)
data = np.random.normal(100, 20, 1000)

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(8, 4))

# Create the histogram
ax.hist(data, bins=30, edgecolor='black', color='skyblue')

# Add title and labels
ax.set_title("Basic Histogram")
ax.set_xlabel("Values")
ax.set_ylabel("Frequency")

plt.show()
```

In this example, we:
1. Generate random data from a normal distribution using NumPy.
2. Create a figure and axis object using `plt.subplots()`.
3. Use `ax.hist()` to create the histogram.
4. Set the title and axis labels.


### <a id='toc4_2_'></a>[Customizing Histograms](#toc0_)


Histograms can be customized to better represent your data and enhance their visual appeal:


```python
# Create two datasets
data1 = np.random.normal(100, 20, 1000)
data2 = np.random.normal(80, 30, 1000)

fig, ax = plt.subplots(figsize=(8, 4))

# Create overlapping histograms
ax.hist(data1, bins=30, alpha=0.7, label='Dataset 1', edgecolor='black')
ax.hist(data2, bins=30, alpha=0.7, label='Dataset 2', edgecolor='black')

# Customize the plot
ax.set_title("Customized Histogram")
ax.set_xlabel("Values")
ax.set_ylabel("Frequency")
ax.legend()

# Add a grid
ax.grid(True, linestyle='--', alpha=0.7)

# Add vertical lines for means
ax.axvline(data1.mean(), color='r', linestyle='dashed', linewidth=2, label='Mean 1')
ax.axvline(data2.mean(), color='g', linestyle='dashed', linewidth=2, label='Mean 2')

# Update legend
ax.legend()
```

This example demonstrates:
- Creating overlapping histograms to compare distributions
- Customizing transparency with `alpha`
- Adding a legend
- Including a grid for easier value comparison
- Adding vertical lines to indicate means
- Adjusting the overall layout and appearance


### <a id='toc4_3_'></a>[When to Use Histograms](#toc0_)


Histograms are ideal when you want to:
- Visualize the distribution of a single continuous variable
- Identify the central tendency, spread, and shape of the distribution
- Detect outliers or unusual patterns in the data
- Compare distributions of different datasets


### <a id='toc4_4_'></a>[Histogram Variations and Related Plots](#toc0_)


1. **Kernel Density Estimation (KDE) plots**: Smooth version of the histogram
2. **Cumulative histogram**: Shows the cumulative counts or percentages
3. **2D histograms**: For visualizing the joint distribution of two variables


```python
# Example of a KDE plot
from scipy import stats

fig, ax = plt.subplots(figsize=(8, 4))

# Plot histogram and KDE
ax.hist(data1, bins=30, density=True, alpha=0.7, edgecolor='black')
kde = stats.gaussian_kde(data1)
x_range = np.linspace(data1.min(), data1.max(), 100)
ax.plot(x_range, kde(x_range), label='KDE')

ax.set_title("Histogram with KDE")
ax.set_xlabel("Values")
ax.set_ylabel("Density")
ax.legend()
```

This example shows how to overlay a Kernel Density Estimation plot on a histogram, providing a smooth estimate of the probability density function of the data.


> Kernel Density Estimation (KDE) is a non-parametric method for estimating the probability density function of a random variable based on a finite data sample. In simple terms, it's a way to smooth out a histogram to create a continuous probability distribution.

Understanding how to create and interpret histograms is crucial for exploratory data analysis and for communicating the distribution of your data effectively.

## <a id='toc5_'></a>[Combining Multiple Plot Types](#toc0_)

Combining different plot types can provide a more comprehensive view of your data, allowing you to showcase various aspects of your dataset in a single figure. Matplotlib's flexibility enables us to create complex, multi-faceted visualizations easily.


### <a id='toc5_1_'></a>[Creating a Figure with Multiple Subplots](#toc0_)


Let's start by creating a figure that combines different plot types:


```python
# Create data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 35, 14, 27, 10]
data = np.random.normal(100, 20, 1000)

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Line plot
axs[0, 0].plot(x, y1, label='sin(x)')
axs[0, 0].plot(x, y2, label='cos(x)')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(y1, y2)
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('sin(x)')
axs[0, 1].set_ylabel('cos(x)')

# Bar chart
axs[1, 0].bar(categories, values)
axs[1, 0].set_title('Bar Chart')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')

# Histogram
axs[1, 1].hist(data, bins=30, edgecolor='black')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Values')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout and display
plt.tight_layout()
```

In this example, we:
1. Create a 2x2 grid of subplots using `plt.subplots(2, 2)`.
2. Plot different types of charts in each subplot.
3. Customize each subplot with titles and labels.
4. Use `plt.tight_layout()` to automatically adjust spacing between subplots.


### <a id='toc5_2_'></a>[Combining Plot Types on the Same Axes](#toc0_)


Sometimes, it's useful to combine different plot types on the same axes:


```python
# Create data
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/10)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Line plot
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

# Scatter plot
ax.scatter(x[::3], y3[::3], color='red', s=50, label='exp(-x/10)')

# Bar plot for y3
ax.bar(x[::3], y3[::3], alpha=0.3, width=0.2, color='green', label='exp(-x/10) bars')

# Customize the plot
ax.set_title('Combination of Plot Types')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

This example demonstrates:
- Plotting multiple line plots on the same axes
- Adding a scatter plot to show specific points
- Including a bar plot to emphasize certain values
- Customizing the overall appearance with a legend and grid


### <a id='toc5_3_'></a>[When to Combine Plot Types](#toc0_)


Combining plot types can be beneficial when you want to:
- Show different aspects of the same data
- Compare different datasets or variables
- Highlight specific data points or ranges
- Create a dashboard-like view of your data


### <a id='toc5_4_'></a>[Tips for Combining Plot Types](#toc0_)


1. **Keep it clear**: Don't overcrowd your plot. Make sure each element serves a purpose.
2. **Use consistent colors**: If the same data appears in multiple forms, use the same color for easy identification.
3. **Leverage secondary y-axes**: For data with different scales, consider using a secondary y-axis.
4. **Add explanatory text**: Use annotations to explain complex parts of your visualization.
5. **Consider interactivity**: For complex combinations, interactive plots (e.g., using libraries like Plotly) can be very effective.


Combining plot types effectively can lead to powerful, insightful visualizations. However, it's important to balance complexity with clarity to ensure your message is effectively communicated.

### <a id='toc6_1_'></a>[Practical Example 1: Analyzing Stock Prices](#toc0_)


Let's create a visualization that combines a line plot of stock prices with a bar chart of trading volumes.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create sample data
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
prices = 100 + np.cumsum(np.random.randn(len(dates)) * 0.25)
volumes = np.random.randint(1000000, 5000000, size=len(dates))

# Create figure and axis objects
fig, ax1 = plt.subplots(figsize=(8, 3))

# Plot stock prices
ax1.plot(dates, prices, color='blue', label='Stock Price')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price ($)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis for volume
ax2 = ax1.twinx()
ax2.bar(dates, volumes, alpha=0.3, color='gray', label='Trading Volume')
ax2.set_ylabel('Volume', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

# Customize the plot
plt.title('Stock Price and Trading Volume Over Time')
fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
```

This example demonstrates how to:
- Combine a line plot (for stock prices) with a bar chart (for trading volumes)
- Use dual y-axes to display different scales
- Customize colors and labels for clarity


### <a id='toc6_2_'></a>[Practical Example 2: Analyzing Survey Results](#toc0_)


Let's visualize the results of a hypothetical survey about preferred programming languages.


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
languages = ['Python', 'JavaScript', 'Java', 'C++', 'Ruby']
popularity = [68, 59, 44, 36, 23]
difficulty = [3, 4, 7, 8, 5]  # on a scale of 1-10

# Create figure and axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

# Bar chart for popularity
ax1.bar(languages, popularity, color='skyblue')
ax1.set_title('Programming Language Popularity')
ax1.set_xlabel('Languages')
ax1.set_ylabel('Popularity (%)')
for i, v in enumerate(popularity):
    ax1.text(i, v, str(v), ha='center', va='bottom')

# Scatter plot for popularity vs difficulty
ax2.scatter(difficulty, popularity, s=100)
for i, lang in enumerate(languages):
    ax2.annotate(lang, (difficulty[i], popularity[i]), xytext=(5, 5), 
                 textcoords='offset points')
ax2.set_title('Popularity vs. Perceived Difficulty')
ax2.set_xlabel('Difficulty (1-10)')
ax2.set_ylabel('Popularity (%)')

plt.tight_layout()
```

This example shows how to:
- Create a bar chart to show the popularity of programming languages
- Use a scatter plot to explore the relationship between popularity and perceived difficulty
- Add annotations to make the plots more informative
