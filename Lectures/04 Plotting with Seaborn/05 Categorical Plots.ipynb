# Categorical Plots in Seaborn

Categorical plots are powerful tools for visualizing relationships between variables when one or more of those variables are categorical (i.e., divided into discrete groups). Seaborn, a statistical data visualization library built on top of Matplotlib, provides a rich set of functions for creating these plots easily and effectively.


Categorical plots are visualizations that show the distribution of a quantitative variable across different categories. They are particularly useful when you want to compare values between different groups or analyze the distribution of data within specific categories.


ðŸ”‘ **Key Concept:** Categorical plots help us understand how a numerical variable behaves across different categories or groups.


Categorical plots serve several important purposes in data analysis and visualization:

1. **Comparison:** They allow for easy comparison of values across different categories.
2. **Distribution:** They show the distribution of data within each category.
3. **Outlier Detection:** Many categorical plots help in identifying outliers within categories.
4. **Trend Analysis:** Some categorical plots can reveal trends across ordered categories.


Understanding the relationship between categorical and numerical variables is crucial in many fields, from business analytics to scientific research.


Seaborn offers various types of categorical plots, which can be broadly categorized into three families:

1. **Categorical Scatterplots:**
   - Strip plots
   - Swarm plots

2. **Categorical Distribution Plots:**
   - Box plots
   - Violin plots
   - Boxen plots

3. **Categorical Estimate Plots:**
   - Point plots
   - Bar plots
   - Count plots

Each type of plot has its strengths and is suited for different kinds of data and analysis goals.


Seaborn provides a unified interface for creating categorical plots through the `catplot()` function. This function acts as a figure-level interface to many of the axes-level categorical plot functions.


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Example usage of catplot
sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="day")
```

Using `catplot()` allows for easy switching between different plot types by changing the `kind` parameter, making it simple to explore your data from different perspectives.


Selecting the appropriate type of categorical plot depends on several factors:

- The nature of your data
- The specific questions you're trying to answer
- The level of detail you want to show
- The size of your dataset


There's often no single "best" plot type. It's common to use multiple types of categorical plots to gain a comprehensive understanding of your data. In the following sections, we'll explore each type of categorical plot in detail, learning how to create them, when to use them, and how to interpret the results.

## <a id='toc1_'></a>[Categorical Scatterplots](#toc0_)

Categorical scatterplots are an excellent way to visualize the distribution of individual data points across different categories. Seaborn offers two main types of categorical scatterplots: strip plots and swarm plots. These plots are particularly useful when you want to show all individual observations while also revealing the underlying distribution of the data.


### <a id='toc1_1_'></a>[Strip Plots](#toc0_)


Strip plots are a simple yet effective way to visualize the distribution of data points across categories.


Let's create a strip plot using a dataset of penguin measurements:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a strip plot
sns.stripplot(
    data=penguins, x="species", y="body_mass_g", jitter=True, palette="viridis"
)
plt.title("Body Mass Distribution of Penguin Species")
plt.xlabel("Penguin Species")
plt.ylabel("Body Mass (g)")
```

This plot shows the distribution of body mass for different penguin species. Each point represents an individual penguin.


ðŸ’¡ **Note:** The `jitter` parameter adds a small amount of random noise to the x-axis position of each point, helping to reduce overplotting.


Use strip plots when:
- You want to show all individual data points.
- The dataset is relatively small to medium-sized.
- You need to quickly compare distributions across categories.

Strip plots are less effective when:
- The dataset is very large, leading to significant overplotting.
- You need to precisely compare central tendencies or spread across categories.


We can add more dimensions to our strip plot by using the `hue` parameter:


```python
sns.stripplot(
    data=penguins, x="species", y="body_mass_g", hue="sex", dodge=True, palette="Set2"
)
plt.title("Body Mass Distribution of Penguin Species by Sex")
plt.xlabel("Penguin Species")
plt.ylabel("Body Mass (g)")
plt.legend(title="Sex", loc="upper right")
```

This enhanced plot now shows the body mass distribution for each species, split by sex.


### <a id='toc1_2_'></a>[Swarm Plots](#toc0_)


Swarm plots are an alternative to strip plots that adjust the positions of points to avoid overlap.


Let's create a swarm plot using the same penguins dataset:


```python
sns.swarmplot(data=penguins, x="species", y="flipper_length_mm")
plt.title("Flipper Length Distribution of Penguin Species")
plt.xlabel("Penguin Species")
plt.ylabel("Flipper Length (mm)")
```

This plot shows the distribution of flipper lengths for different penguin species, with points adjusted to avoid overlap.


Use swarm plots when:
- You want to show all individual data points without overlap.
- The dataset is small to medium-sized.
- You need to visualize the density of data points within categories.

Swarm plots are less effective when:
- The dataset is very large, as they don't scale well to large numbers of observations.
- You need to preserve exact y-axis values for each point.


Swarm plots can be effectively combined with other plot types for a more comprehensive view:


```python
sns.boxplot(data=penguins, x="species", y="flipper_length_mm", color="lightyellow")
sns.swarmplot(data=penguins, x="species", y="flipper_length_mm")
plt.title("Flipper Length Distribution of Penguin Species (Box Plot with Swarm)")
plt.xlabel("Penguin Species")
plt.ylabel("Flipper Length (mm)")
```

This combined plot shows both the overall distribution (box plot) and individual data points (swarm plot). Combining plot types can provide a more nuanced view of your data, showing both summary statistics and individual observations simultaneously.


Categorical scatterplots, whether strip plots or swarm plots, are powerful tools for visualizing the distribution of data across categories. They allow you to see individual data points, making them excellent for smaller datasets or when you need to identify outliers. However, they can become cluttered with large datasets, in which case you might consider using other types of categorical plots that we'll explore in the following sections.


â—ï¸ **Important Note:** Always consider the size of your dataset and the specific insights you're trying to gain when choosing between strip plots and swarm plots, or when deciding to use a categorical scatterplot at all.

## <a id='toc2_'></a>[Categorical Distribution Plots](#toc0_)

Categorical distribution plots are powerful tools for visualizing the distribution of a quantitative variable across different categories. Seaborn offers three main types of categorical distribution plots: box plots, violin plots, and boxen plots. Each of these plot types provides a unique perspective on the data distribution, allowing for rich insights into your dataset.


### <a id='toc2_1_'></a>[Box Plots](#toc0_)


Box plots, also known as box-and-whisker plots, are a classic way to summarize the distribution of data.


Let's create a box plot using a dataset of diamond characteristics:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a box plot
sns.boxplot(data=diamonds, x="cut", y="price", hue="cut")
plt.title("Diamond Prices by Cut Quality")
plt.xlabel("Cut Quality")
plt.ylabel("Price (USD)")
```

This plot shows the distribution of diamond prices for each cut quality category. The box represents the interquartile range (IQR), the line inside the box is the median, and the whiskers extend to show the rest of the distribution, excluding outliers.


Use box plots when:
- You want to show the median, quartiles, and potential outliers.
- You need to compare distributions across multiple categories.
- The dataset is large and you want a concise summary.

Box plots are less effective when:
- The distribution is highly skewed or multimodal.
- You need to show the exact shape of the distribution.


We can add more dimensions to our box plot by using the `hue` parameter:


```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=diamonds, x="cut", y="price", hue="color", palette="Set3", gap=0.2)
plt.title("Diamond Prices by Cut Quality and Color")
plt.xlabel("Cut Quality")
plt.ylabel("Price (USD)")
plt.legend(title="Diamond Color", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
```

This enhanced plot now shows the price distribution for each cut quality, further divided by diamond color.


### <a id='toc2_2_'></a>[Violin Plots](#toc0_)


Violin plots combine the features of box plots with kernel density estimation, providing a richer view of the data distribution.


Let's create a violin plot using the same diamonds dataset:


```python
sns.violinplot(data=diamonds, x="cut", y="price", hue="cut")
plt.title("Distribution of Diamond Prices by Cut Quality")
plt.xlabel("Cut Quality")
plt.ylabel("Price (USD)")
```

This plot shows the full distribution of prices for each cut quality, with wider sections representing a higher probability of observations in that price range.


Use violin plots when:
- You want to show the full distribution of the data.
- The dataset has complex distributions (e.g., bimodal).
- You need to compare distributions across categories visually.

Violin plots are less effective when:
- The dataset is very small.
- You need to read exact values or quartiles easily.


We can customize violin plots to show inner quartiles and split them by a categorical variable:


```python
plt.figure(figsize=(14, 7))
sns.violinplot(
    data=diamonds,
    x="cut",
    y="price",
    hue="clarity",
    split=True,
    inner="quartile",
    palette="Set2",
)
plt.title("Diamond Price Distribution by Cut and Clarity")
plt.xlabel("Cut Quality")
plt.ylabel("Price (USD)")
plt.legend(title="Clarity", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
```

This enhanced plot shows the price distribution for each cut quality, split by clarity, with quartiles marked inside each violin.


### <a id='toc2_3_'></a>[Boxen Plots](#toc0_)


Boxen plots, also known as letter-value plots, are an extension of box plots that show more quantiles, providing a more detailed view of the distribution.


Let's create a boxen plot using our diamonds dataset:


```python
sns.boxenplot(data=diamonds, x="cut", y="price", hue="cut")
plt.title("Detailed Distribution of Diamond Prices by Cut Quality")
plt.xlabel("Cut Quality")
plt.ylabel("Price (USD)")
```

This plot shows a more detailed distribution of prices for each cut quality, with multiple boxes representing different quantiles.


Use boxen plots when:
- You have a large dataset and want to show more detail than a standard box plot.
- You need to visualize the distribution shape, especially in the tails.
- You want to compare distributions across many categories.

Boxen plots are less effective when:
- The dataset is small (fewer than 1000 points per category).
- You need a simple, easily interpretable visualization for non-technical audiences.


We can further customize boxen plots to show more levels and add a hue dimension:


```python
plt.figure(figsize=(14, 7))
sns.boxenplot(
    data=diamonds,
    x="cut",
    y="price",
    hue="clarity",
    palette="YlOrBr",
    k_depth="tukey",
    linewidth=0.5,
    gap=0.2,
)
plt.title("Detailed Diamond Price Distribution by Cut and Clarity")
plt.xlabel("Cut Quality")
plt.ylabel("Price (USD)")
plt.legend(title="Clarity", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
```

This enhanced plot provides a highly detailed view of the price distribution across cut qualities and clarity levels.


Categorical distribution plots offer powerful ways to visualize and compare distributions across categories. Box plots provide a concise summary, violin plots show the full shape of the distribution, and boxen plots offer a detailed view for large datasets.


Choosing the right plot type can significantly impact how effectively you communicate your data insights. Each plot type has its strengths, and the best choice depends on your specific dataset and the story you want to tell with your data.


â—ï¸ **Important Note:** Always consider your audience and the complexity of your data when choosing between these plot types. While boxen plots offer the most detail, they might be overwhelming for non-technical audiences. In such cases, simpler box plots or violin plots might be more effective.

## <a id='toc3_'></a>[Categorical Estimate Plots](#toc0_)

Categorical estimate plots are powerful tools for visualizing summary statistics of a quantitative variable across different categories. Seaborn offers three main types of categorical estimate plots: point plots, bar plots, and count plots. Each of these plot types provides a unique way to represent central tendencies and comparisons across categories.


### <a id='toc3_1_'></a>[Point Plots](#toc0_)


Point plots are excellent for showing the central tendency of a numeric variable for different levels of categorical variables.


Let's create a point plot using a dataset of penguin measurements:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a point plot
sns.pointplot(
    data=penguins, x="island", y="body_mass_g", hue="sex", capsize=0.1, palette="Set2"
)
plt.title("Average Body Mass of Penguins by Island and Sex")
plt.xlabel("Island")
plt.ylabel("Body Mass (g)")
plt.legend(title="Sex", loc="upper right")
```

This plot shows the average body mass of penguins for each island, split by sex. The lines connecting points for each hue level make it easy to see interactions between variables.


Use point plots when:
- You want to focus on comparing levels of categorical variables.
- You need to show interactions between two categorical variables.
- You want to emphasize differences in slopes rather than absolute values.

Point plots are less effective when:
- You need to show the full distribution of the data.
- The dataset is very small or has many outliers.


We can customize point plots to show different error bars and use markers for better accessibility:


```python
sns.pointplot(
    data=penguins,
    x="island",
    y="body_mass_g",
    hue="species",
    markers=["o", "s", "D"],
    linestyles=["-", "--", "-."],
    capsize=0.2,
    errorbar=("ci", 95),
    palette="deep",
)
plt.title("Average Body Mass of Penguin Species by Island")
plt.xlabel("Island")
plt.ylabel("Body Mass (g)")
plt.legend(title="Species", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
```

This enhanced plot uses different markers and line styles for each species, making it more accessible and informative.


### <a id='toc3_2_'></a>[Bar Plots](#toc0_)


Bar plots are a classic way to show comparisons between categories using the height of the bars.


Let's create a bar plot using our penguins dataset:


```python
ax = sns.barplot(
    data=penguins,
    x="species",
    y="flipper_length_mm",
    errorbar=("ci", 95),
    capsize=0.1,
    hue="sex",
)
ax.legend(title="Sex", loc="upper right", bbox_to_anchor=(1.25, 1))
plt.title("Average Flipper Length by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
```

This plot shows the average flipper length for each penguin species with 95% confidence intervals.


Use bar plots when:
- You want to compare quantities across different categories.
- The zero point is meaningful for your data.
- You need to show aggregated values (like means or sums) for each category.

Bar plots are less effective when:
- The zero point is not meaningful for your data.
- You need to show the distribution of the data within categories.


### <a id='toc3_3_'></a>[Count Plots](#toc0_)


Count plots are used to show the frequency of observations in each categorical bin.


Let's create a count plot using the Titanic dataset:


```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

sns.countplot(data=titanic, x="class", hue="survived", palette="coolwarm")
plt.title("Passenger Count by Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
```

This plot shows the count of passengers in each class, split by survival status.


Use count plots when:
- You want to show the distribution of a categorical variable.
- You need to compare the frequency of different categories.
- You're dealing with nominal (unordered) categorical data.

Count plots are less effective when:
- You're working with continuous data.
- You need to show relationships between multiple variables.


We can customize count plots to show percentages instead of raw counts:


```python
sns.countplot(
    data=titanic,
    y="embark_town",
    hue="class",
    stat="percent",
    palette="Set2",
    order=titanic["embark_town"].value_counts().index,
)
plt.title("Percentage of Passengers by Embarkation Town and Class")
plt.xlabel("Percentage")
plt.ylabel("Embarkation Town")
plt.legend(title="Class", loc="lower right")
```

This enhanced plot shows the percentage of passengers from each embarkation town, split by passenger class.


Categorical estimate plots offer powerful ways to visualize summary statistics and comparisons across categories. Point plots excel at showing interactions between variables, bar plots are great for comparing aggregated values, and count plots are perfect for showing the distribution of categorical variables.


Choosing the right plot type can significantly impact how effectively you communicate your data insights. Each plot type has its strengths, and the best choice depends on your specific dataset and the story you want to tell with your data.


â—ï¸ **Important Note:** Always consider the nature of your data and your audience when choosing between these plot types. While bar plots are widely recognized, point plots might be more effective for showing complex interactions. Count plots are excellent for quickly conveying the distribution of categorical data.

## <a id='toc4_'></a>[Combining Categorical Plots for Enhanced Insights](#toc0_)

Combining different types of categorical plots can provide a more comprehensive view of your data, allowing for richer insights and more nuanced interpretations. By layering multiple plot types, we can leverage the strengths of each to create powerful, informative visualizations.


Combining plots allows you to:
1. Show both overall trends and individual data points
2. Compare different statistical measures simultaneously
3. Highlight multiple aspects of your data in a single visualization
4. Create more engaging and informative graphics


Let's explore some effective combinations and see how they can enhance our understanding of the data.


### <a id='toc4_1_'></a>[Box Plots with Strip Plots](#toc0_)


Combining box plots with strip plots allows us to see both the distribution summary and individual data points.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

sns.boxplot(data=tips, x="day", y="total_bill", color="lightgray")
sns.stripplot(
    data=tips, x="day", y="total_bill", color="darkblue", alpha=0.5, jitter=True
)

plt.title("Distribution of Total Bill by Day with Individual Data Points")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
```

This combination allows you to see both the overall distribution and any potential outliers or patterns in the raw data.


### <a id='toc4_2_'></a>[Violin Plots with Swarm Plots](#toc0_)


Violin plots show the distribution shape, while swarm plots add individual data points without overlap.


```python
sns.violinplot(data=tips, x="day", y="total_bill", inner=None, color="lightblue")
sns.swarmplot(data=tips, x="day", y="total_bill", color="darkblue", alpha=0.7)

plt.title("Distribution of Total Bill by Day with Swarm Plot")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
```

This combination provides a detailed view of the data distribution while still showing each individual data point, making it easier to spot clusters and outliers.


### <a id='toc4_3_'></a>[Bar Plots with Point Plots](#toc0_)


Combining bar plots with point plots can show both the mean values and the trend across categories.


```python
sns.barplot(data=tips, x="day", y="total_bill", alpha=0.8, color="lightblue")
sns.pointplot(
    data=tips, x="day", y="total_bill", color="darkblue", markers="D", linestyles="--"
)

plt.title("Average Total Bill by Day with Trend")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
```

This combination is particularly useful when you want to emphasize both the individual category values and the overall trend across categories.


### <a id='toc4_4_'></a>[Box Plots with Violin Plots](#toc0_)


Combining box plots with violin plots provides a comprehensive view of the data distribution.


```python
sns.violinplot(data=tips, x="day", y="total_bill", inner=None, color="lightsalmon")
sns.boxplot(data=tips, x="day", y="total_bill", width=0.3, color="white")

plt.title("Distribution of Total Bill by Day (Violin + Box Plot)")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
```

This combination shows the full distribution shape from the violin plot along with key statistics (median, quartiles, and potential outliers) from the box plot.


Combining different types of categorical plots can significantly enhance the insights we gain from our data. By layering multiple plot types, we can:

1. Show both summary statistics and individual data points
2. Highlight different aspects of the data distribution
3. Reveal trends and patterns that might not be apparent with a single plot type
4. Create more engaging and informative visualizations


When combining plots, always consider the story you want to tell with your data and choose combinations that complement each other to provide a fuller picture.


While combined plots can be very informative, be careful not to overcrowd your visualization. Always strive for clarity and ease of interpretation. By mastering the art of combining categorical plots, you can create powerful, insightful visualizations that effectively communicate complex data relationships to your audience.

## <a id='toc5_'></a>[Figure-Level Interface: Catplot](#toc0_)

The `catplot()` function in Seaborn provides a powerful, flexible interface for creating categorical plots. It's a figure-level function that allows you to create various types of categorical plots on a `FacetGrid`, offering a consistent API for different plot types and easy creation of multi-panel figures.


`catplot()` is a versatile function that can create different types of categorical plots, including strip plots, box plots, violin plots, and more, all using a consistent interface.


Use `catplot()` when:
- You want to create categorical plots with a consistent API
- You need to create multi-panel figures (facets) easily
- You want to switch between different plot types quickly


Let's start with a basic example using the Titanic dataset:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a basic catplot (default is a strip plot)
sns.catplot(data=titanic, x="class", y="age")
plt.title("Distribution of Age by Passenger Class")
```

This creates a strip plot showing the distribution of passenger ages for each class. The default plot type for `catplot()` is a strip plot, but you can easily change this using the `kind` parameter.


### <a id='toc5_1_'></a>[Changing Plot Types](#toc0_)


One of the strengths of `catplot()` is the ability to easily switch between different plot types:


```python
# Create a box plot
sns.catplot(data=titanic, x="class", y="age", kind="box")
plt.title("Age Distribution by Passenger Class (Box Plot)")
```

```python
# Create a violin plot
sns.catplot(data=titanic, x="class", y="age", kind="violin")
plt.title("Age Distribution by Passenger Class (Violin Plot)")
```

This flexibility allows you to quickly explore different representations of your data without changing your code structure significantly.


### <a id='toc5_2_'></a>[Adding a Hue Dimension](#toc0_)


`catplot()` makes it easy to add another categorical dimension using the `hue` parameter:


```python
sns.catplot(data=titanic, x="class", y="age", hue="sex", kind="box")
plt.title("Age Distribution by Passenger Class and Sex")
```

This plot now shows the age distribution for each class, further split by sex.


### <a id='toc5_3_'></a>[Creating Faceted Plots](#toc0_)


One of the most powerful features of `catplot()` is the ability to create faceted plots easily:


```python
g = sns.catplot(
    data=titanic, x="class", y="survived", col="sex", kind="bar", height=4, aspect=0.7
)
g.set_axis_labels("Passenger Class", "Survival Rate")
g.set_titles("{col_name}")
```

This creates separate bar plots for male and female passengers, showing survival rates by class.


â—ï¸ **Important Note:** The `catplot()` function returns a `FacetGrid` object, which you can further customize using its methods.


### <a id='toc5_4_'></a>[Customizing Catplots](#toc0_)


You can pass additional arguments to customize your plots:


```python
sns.catplot(
    data=titanic,
    x="class",
    y="age",
    hue="sex",
    kind="violin",
    split=True,
    inner="stick",
    palette="Set3",
    height=6,
    aspect=1.2,
)
plt.title("Age Distribution by Class and Sex (Split Violin Plot)")
```

This creates a split violin plot with inner "stick" representations of individual data points.


### <a id='toc5_5_'></a>[Combining Multiple Plot Types](#toc0_)


You can layer different plot types on top of each other for more comprehensive visualizations:


```python
g = sns.catplot(data=titanic, x="class", y="age", kind="violin", inner=None, color=".8")
sns.swarmplot(data=titanic, x="class", y="age", color="k", size=2, ax=g.ax)
g.set_titles("Age Distribution by Class with Individual Points")
```

This combines a violin plot to show the overall distribution with a swarm plot to show individual data points.


The `catplot()` function in Seaborn provides a powerful and flexible way to create categorical plots. Its key advantages include:

1. Consistent API across different plot types
2. Easy creation of faceted plots
3. Built-in legend placement and other sensible defaults
4. Returned `FacetGrid` object for further customization


ðŸ”‘ **Key Takeaway:** `catplot()` is an excellent tool for exploratory data analysis and creating publication-quality figures, allowing you to quickly switch between different categorical plot types and easily create multi-panel visualizations.


By mastering `catplot()`, you'll be able to create a wide variety of categorical plots efficiently, making it easier to explore and communicate insights from your data.

- Consider the size of your dataset when selecting between strip plots and swarm plots.
- Use distribution plots (box, violin, boxen) to show the full shape of data distribution across categories.
- Estimate plots are useful for comparing central tendencies or frequencies across categories.
- Combining plot types can provide a more nuanced view of your data.
- The `catplot()` function offers a flexible interface for creating and customizing categorical plots.


Remember to always consider your audience and the complexity of your data when choosing between these plot types. While some plots offer more detail, simpler visualizations might be more effective for non-technical audiences.
