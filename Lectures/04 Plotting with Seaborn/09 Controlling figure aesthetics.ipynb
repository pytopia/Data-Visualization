## <a id='toc1_'></a>[Seaborn Figure Styles](#toc0_)

Seaborn offers a variety of predefined styles that can dramatically change the appearance of your plots. These styles are designed to suit different types of data, presentation contexts, and personal preferences. Seaborn's figure styles control the overall look and feel of your plots, including background color, grid lines, and other aesthetic elements.


Seaborn provides five main style themes:

1. **darkgrid** (default)
2. **whitegrid**
3. **dark**
4. **white**
5. **ticks**


Let's explore each of these styles and their use cases.


### <a id='toc1_1_'></a>[Setting the Style](#toc0_)


To set a style, use the `set_style()` function:


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set the style
sns.set_style("whitegrid")

# Create a sample plot
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sinplot()
```

### <a id='toc1_2_'></a>[Exploring Different Styles](#toc0_)


**1. Darkgrid Style**


```python
sns.set_style("darkgrid")
sinplot()
plt.title("Darkgrid Style")
```

The darkgrid style is excellent for plots where you want to emphasize the data while still providing reference lines.


**2. Whitegrid Style**


```python
sns.set_style("whitegrid")
sinplot()
plt.title("Whitegrid Style")
```

Whitegrid is ideal for plots with dense data, as it provides a clean background that doesn't compete with the data.


**3. Dark Style**


```python
sns.set_style("dark")
sinplot()
plt.title("Dark Style")
```

The dark style is great for presentations or when you want the data to stand out against a dark background.


**4. White Style**


```python
sns.set_style("white")
sinplot()
plt.title("White Style")
```

White style provides a clean, minimalist look that's perfect for publication-quality figures.


**5. Ticks Style**


```python
sns.set_style("ticks")
sinplot()
plt.title("Ticks Style")
```

The ticks style adds tick marks to the axes, providing more precise reference points.


ðŸ’¡ **Pro Tip:** Choose your style based on the nature of your data and the context of your presentation. For example, use whitegrid for busy plots and dark for emphasis in presentations.


### <a id='toc1_3_'></a>[Removing Spines](#toc0_)


Seaborn allows you to easily remove unnecessary spines (the lines on the edges of the plot) using the `despine()` function:


```python
sns.set_style("ticks")
sinplot()
sns.despine(left=True)
plt.title("Ticks Style with Despine")
```

This can help reduce visual clutter and focus attention on the data.


### <a id='toc1_4_'></a>[Temporarily Setting Styles](#toc0_)


You can temporarily set a style using a context manager:


```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()

plt.subplot(212)
sinplot(-1)
```

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])
    sinplot(6)

with sns.axes_style("white"):
    ax = f.add_subplot(gs[0, 1])
    sinplot(6)

with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[1, 0])
    sinplot(6)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(gs[1, 1])
    sinplot(6)

f.tight_layout()
```

This allows you to create figures with differently-styled subplots.


ðŸ¤” **Why This Matters:** The ability to easily switch between styles and customize spine visibility gives you fine-grained control over your plot's appearance, allowing you to create visualizations that best suit your data and audience.


By mastering Seaborn's figure styles, you can create visually appealing and professional-looking plots that effectively communicate your data insights.

## <a id='toc2_'></a>[Customizing Plot Elements](#toc0_)

While Seaborn's predefined styles offer great out-of-the-box aesthetics, you often need to fine-tune specific elements of your plots. Seaborn provides various ways to customize individual plot components, allowing you to create visualizations that perfectly match your needs.


Customizing plot elements in Seaborn involves modifying specific aspects of your visualization, such as colors, fonts, and line styles, to enhance clarity and visual appeal.


Seaborn allows you to override specific style parameters using dictionaries:


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Override style parameters
sns.set_style(
    "darkgrid", {"axes.facecolor": ".9", "grid.color": ".6", "grid.linestyle": ":"}
)

# Example plot
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title("Customized Dark Grid")
```

This approach lets you tweak individual style elements while maintaining the overall theme.


### <a id='toc2_1_'></a>[Customizing Color Palettes](#toc0_)


Seaborn offers a rich set of color palette functions to customize the colors in your plots:


```python
# Set a custom color palette
sns.set_palette("husl", 8)

# Example plot with multiple lines
for i in range(8):
    plt.plot(x, np.sin(x - i), label=f"Line {i+1}")
plt.legend()
plt.title("Custom Color Palette")
```

ðŸ’¡ **Pro Tip:** You can create your own color palettes using `sns.color_palette()` for even more control over your plot's appearance. We will see more about this in the upcoming lectures.


### <a id='toc2_2_'></a>[Modifying Plot-Specific Elements](#toc0_)


Many Seaborn plotting functions allow direct customization of plot elements:


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
df = pd.DataFrame(
    {
        "x": np.random.rand(100),
        "y": np.random.rand(100),
        "size": np.random.rand(100) * 100,
    }
)

# Customized scatter plot
sns.scatterplot(
    data=df,
    x="x",
    y="y",
    size="size",
    sizes=(20, 200),
    hue="size",
    palette="viridis",
    legend="brief",  # or "full"
)
plt.title("Customized Scatter Plot")
```

This example demonstrates customizing marker sizes, colors, and legend appearance in a scatter plot.


### <a id='toc2_3_'></a>[Adjusting Text Elements](#toc0_)


Customizing text elements like titles, labels, and tick marks can greatly enhance your plot's readability:


```python
plt.figure(figsize=(6, 4))
sns.lineplot(x=x, y=np.sin(x))
plt.title("Sine Wave", fontsize=20, fontweight='bold')
plt.xlabel("X-axis", fontsize=14)
plt.ylabel("Y-axis", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12);
```

### <a id='toc2_4_'></a>[Using Matplotlib's Customization Options](#toc0_)


Remember that Seaborn is built on top of Matplotlib, so you can always fall back on Matplotlib's extensive customization options:


```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.lineplot(x=x, y=np.sin(x), ax=ax)
ax.set_title("Customized Axes", fontsize=20)
ax.set_xlabel("X-axis", fontsize=14)
ax.set_ylabel("Y-axis", fontsize=14)
ax.grid(True, linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

This example demonstrates customizing the axes, grid, and spines using Matplotlib functions.


The ability to fine-tune individual plot elements allows you to create visualizations that not only look appealing but also effectively highlight the most important aspects of your data.


â—ï¸ **Important Note:** While customization is powerful, it's essential to maintain consistency across your visualizations, especially when presenting multiple plots together.


By mastering these customization techniques, you can create unique, polished, and highly effective data visualizations that perfectly suit your specific needs and preferences.

## <a id='toc3_'></a>[Scaling Plot Elements](#toc0_)

In data visualization, one size doesn't fit all. The context in which you present your plotsâ€”whether it's a research paper, a presentation slide, or a posterâ€”often dictates the ideal size and scale of your visualization elements. Seaborn provides powerful tools to adjust the scale of your plots, ensuring they look great in any setting.


ðŸ”‘ **Key Concept:** Scaling in Seaborn allows you to adjust the size of plot elements such as text, lines, and markers to suit different presentation contexts without changing the underlying data representation.


Seaborn introduces the concept of "plotting contexts" to easily scale plot elements. There are four predefined contexts:

1. `paper`: Smallest elements, suitable for academic papers
2. `notebook`: Default size, good for Jupyter notebooks
3. `talk`: Larger elements, ideal for presentations
4. `poster`: Largest elements, designed for posters


### <a id='toc3_1_'></a>[Setting the Context](#toc0_)


To set a context, use the `set_context()` function:


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.reset_defaults()

def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i))

# Set the context
sns.set_context("talk")

sinplot()
plt.title("Sine Waves (Talk Context)")
plt.show()
```

### <a id='toc3_2_'></a>[Comparing Different Contexts](#toc0_)


Let's visualize how different contexts affect the same plot:


```python
contexts = ['paper', 'notebook', 'talk', 'poster']
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

for i, context in enumerate(contexts):
    sns.set_context(context)
    plt.subplot(2, 2, i+1)
    sinplot()
    plt.title(f"{context.capitalize()} Context")

plt.show()
```

This code creates a 2x2 grid of plots, each using a different context, allowing you to compare the scaling effects side by side.


ðŸ’¡ **Pro Tip:** When preparing visualizations for different mediums, test your plots in various contexts to find the most suitable scaling for your specific use case.


### <a id='toc3_3_'></a>[Fine-tuning Scale Parameters](#toc0_)


While the predefined contexts are useful, you might want more granular control. Seaborn allows you to adjust specific scale parameters:


```python
# Increase font scale while using the 'notebook' context
sns.set_context("notebook", font_scale=1.5)

sinplot()
plt.title("Notebook Context with Increased Font Scale")
plt.show()
```

You can also override specific parameters:


```python
sns.set_context("talk", rc={"lines.linewidth": 2.5})

sinplot()
plt.title("Talk Context with Custom Line Width")
plt.show()
```

### <a id='toc3_4_'></a>[Scaling in Specific Plot Types](#toc0_)


Different types of plots may benefit from different scaling approaches. For example, in a scatter plot with varying point sizes:


```python
# Generate sample data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.random.rand(100) * 300

# Create scatter plot with scaled elements
sns.set_context("poster")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, size=sizes, sizes=(20, 500), legend=False)
plt.title("Scaled Scatter Plot")
plt.show()
```

Here, we use the `poster` context for large overall scaling, and then further customize the point sizes within the plot.


ðŸ¤” **Why This Matters:** Proper scaling ensures that your visualizations are legible and impactful in their intended context, whether it's a dense academic paper or a large conference poster.


### <a id='toc3_5_'></a>[Combining Context with Style](#toc0_)


Remember that you can combine context scaling with style settings for comprehensive control:


```python
sns.set_style("darkgrid")
sns.set_context("talk", font_scale=1.2, rc={"grid.linewidth": 1.5})

sinplot()
plt.title("Dark Grid Style with Talk Context")
plt.show()
```

### <a id='toc3_6_'></a>[Temporary Context Changes](#toc0_)


Like with styles, you can temporarily change the context using a context manager:


```python
with sns.plotting_context("poster"):
    sinplot()
    plt.title("Temporary Poster Context")
plt.show()
```

This is particularly useful when you need different scaling for different parts of your visualization.


â—ï¸ **Important Note:** While scaling can greatly enhance the readability of your plots, be cautious not to overdo it. Extremely large elements can sometimes distort the perception of the data.


By mastering Seaborn's scaling capabilities, you can ensure that your visualizations are not only aesthetically pleasing but also perfectly suited to their presentation context, enhancing the impact and clarity of your data communication.
